```python
def boxplot_iris(feature_names, dataset):
    i = 1
    plt.figure(figsize=(11, 9))
    for col in feature_names:
        plt.subplot(2, 2, i)
        plt.axis("on")
        plt.tick_params(axis="both", left=True)
        dataset[col].plot(kind="box", subplots=True)
        plt.title(col)
        i += 1
    plt.show()

boxplot_iris(iris_df.columns[:-1], iris_df)
```

```python
def histogram_iris(feature_names, dataset):
    i = 1
    plt.figure(figsize=(11, 9))
    for col in feature_names:
        plt.subplot(2, 2, i)
        plt.axis("on")
        plt.tick_params(axis="both", left=True)
        
        dataset[col].hist()
        plt.title(col)
        i += 1
    plt.show()

histogram_iris(iris_df.columns[:-1], iris_df)
```

```python
corr = iris_df.corr()
cmap = sns.diverging_palette(220, 10, as_cmap=True)

plt.figure(figsize=(11, 9))
sns.heatmap(corr, cmap=cmap, vmax=1.0, vmin=-1.0,
           center=0, square=True, linewidths=.5, 
           cbar_kws={"shrink": .5})
plt.show()
```

```python
groups = iris_df.groupby("species")
fig, ax = plt.subplots()

for name, group in groups:
    ax.scatter(group.sepal_length, group.sepal_width,
              marker=".", label=name)

ax.legend()
plt.show()
```

```python
x = iris_df.iloc[:, :-1]
y = iris_df[["species"]]
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                   test_size=0.3,
                                                   random_state=1)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
predict = model.predict(x_test)
```

```python
y_test["predict"] = predict
y_test
```
