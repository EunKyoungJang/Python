```python
from sklearn.datasets import load_breast_cancer
```

```python
# 머신러닝  돈 관련은 거의 회귀분석    나머진 거의 분류
cancer = load_breast_cancer()
cancer
```

```python
# 0:malignant 악성
#  1:benign 양성
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df
```

```python
# columns의단어랑 단어 사이어 공백이 있으면 안된다. 왜요? 불편해서 없애주는게 좋다
df.columns = df.columns.map(lambda x: "_".join(x.split(" ")))
df
# split으로 쪼개고 공백" " 에 "_"로 대체할게, 바꿔줄게  
```

```python
df.info()
```

```python
# 30차원의 x를 가지고 있다. 0-29니까
# 지금은 쓰지만 나중에는 차원을 축소
df["target"] = cancer.target
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(25,15))
for idx, column in enumerate(df.iloc[:, :-1]):
    plt.subplot(6, 5, idx+1)
    sns.histplot(data=df, x=column, hue="target")
    
plt.tight_layout()
plt.show()
```

```python
mal = df[df["target"] == 0]
ben = df[df["target"] == 1]
print(f"양성 : {mal.shape}")
print(f"악성 : {ben.shape}")
```

```python
mal
```

```python
ben
```

```python
# 악성 히트맵
plt.figure(figsize=(25,15))
sns.heatmap(mal.corr(), annot=True)
```

```python
# 양성 히트맵
plt.figure(figsize=(25,15))
sns.heatmap(ben.corr(), annot=True)
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model = DecisionTreeClassifier(max_depth=4) # 최대 4번까지 질문
model.fit(x_train, y_train)
model.score(x_test, y_test)
# max_depth 트리의 깊이
```

```python
from sklearn.tree import export_graphviz
```

```python
'''
impurity : (균일도 / 불순도)
- 규칙 조건 생성에서 중요한 것!
- 정보 균일도가 높은 데이터 셋을 먼저 선택할 수 있도록 규칙 조건을 만드는 것이 중요하다

filled : True로 설정하면 분류
'''

export_graphviz(model, out_file="tree.dot", class_names=["악성", "양성"], feature_names = cancer.feature_names, impurity=False, filled=True)
```
