```python
scores = df["english"]
scores
```

```python
sorted_scores = np.sort(scores)
sorted_scores
```

```python
# median : 중위값
n = len(sorted_scores)
if n % 2 == 0:
    x1 = sorted_scores[n//2-1]
    x2 = sorted_scores[n//2]
    median = (x1 +x2) / 2
else:
    median = sorted_scores[(n+1)//2-1]
median
# 코드가 무슨 내용인지 모르겠으면 찾아보기
```

```python
x = [1, 2, 3, 4, 5]
```

```python
# 모분산
np.var(x)
np.array(x).var()
# 표본분산
pd.Series(x).var()
np.var(x, ddof=1)
# np에서는 ddof default : 0(모집단) = ex) 수능본 사람 전체 
# pd에서는 ddof default : 1(표본)  ex) 전체가 어려우니 교회 수능본 사람들의 데이터를 모집
```

```python
np.std(x)
# std : 표준편차
np.array(x).std()
pd.Series(x).std(ddof=0)
```

```python
men = [72, 74, 77, 68, 66, 75]
women = [45, 48, 52, 53, 46, 50]
```

```python
# 평균
np.mean(men)
np.mean(women)
```

```python
# 표준편차
np.std(men)
np.std(women)
```

```python
# 변동계수 : 편차를 비교하기 위함  # 편차 = 데이터 - 평균   데이터와 평균의 거리  
# 편차가 크다는 말은 데이터가 넓게 펼쳐져 있다 / 편차가 좁다는 데이터가 좁게 모여있다.
np.std(men)/np.mean(men)
np.std(women)/np.mean(women)
```

```python
eng = df["english"]
eng 
```

```python
np.histogram(eng, bins=10, range=(0, 100))
```

```python
freq, _ = np.histogram(eng, bins=10, range=(0, 100))
freq
```

```python
freq_class = [f"{i}~{i+10}"for i in range(0, 100, 10)]
freq_dist_df = pd.DataFrame({"frequency":freq}, index=freq_class)
freq_dist_df
```

```python
# 상대 도수 : 구간 안에 있는 값이 몇개이다.
rel_freq = freq/freq.sum()
rel_freq
# 도수 분포표를 %로 만든 것 
# %로 만든거니까 상대도수의 총합은 1이다.
```

```python
# 누적 상대도수
cum_rel_freq = np.cumsum(rel_freq)
cum_rel_freq
```
