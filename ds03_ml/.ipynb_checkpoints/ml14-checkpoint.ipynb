metrics(평가지표)

MAE(Mean Absolute Error) : 오차 절대값 평균

MSE(Mean Squared Error) : 오차 제곱 평균

RMSE(Root Mean Squared Error) :  "오차 제곱 평균"의 제곱근

R Squared(결정계수) : 예측값 Variance / 실제값 Variance(1에 가까울수록 정확도가 높다)

confusion matrix(혼동 행렬)

- TP (True Positive)
- FP (False Positive)
- TN (True Negative)
- FN (False Negative)


accuracy(score) : 정답 / 전체 -> (TP + TN) / (TP + TN + FP + FN)

정답을 전체로 나눈걸 스코어라고 이야기하고 있다.

f1_score : Precision과 Recall의 조화평균

<!-- 이 아래 내용을 보면 1종 오류와 2종 오류라는 이야기가 있는데, 거기에서 나온 내용이라 통계분석 공부를 해야한다. -->
- 2 * (precision * recall / precision + recall)

- Precision(정밀도) : TP/ TP + TN (True로 분류한 것 중 정답 비율)
- Recall(재현율 / 민감도) : TP / (TP + FN) (실제 True 중 True로 예측한 비율)
- TPR(True Positive Rate) : TP / (TP + FN) (= Recall)
- FRP (False Positive Rate) : FP / (FP + TN) (실제 False 중 False로 예측한 비율)

ROC (Receiver Operating Characteristic) : 재현률을 좀 더 명확하게 판단하기 위함 -> 판별 기준이 낮을수록 Positive로 예측하는 수가 증가한다.

ROC 는 0 ~ 1까지의 그래프로 그려진다(비율이니까)

AUC 는 1인 경우에 가장 분류 모델의 성능이 좋다고 판다.

auc 는 0.5보다 크면 무작위로 만든 분류모델보다는 성능이 좋다고 판단.
        반대로 0.5보다 작으면 분류 모델에 문제가 있다고 판단.
        
책(파이썬 머신러닝 가이드) p.148 ~ 182 에 있다.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, f1_score, accuracy_score, classification_report, precision_recall_curve, roc_curve, roc_auc_score
```

```python
# linear하지 않는 애들은 어떻게 할건데?
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
predict = model.predict(x_test)
predict
```

```python
cm = confusion_matrix(y_test, predict)
cm
# 예측한 값과 실제 값을 배열 형태로 만들어준다
```

```python
ConfusionMatrixDisplay(cm).plot()
# x축은 모델이 예측했다 라는 뜻
# y축은 실제 정답
# 모델은 0이라고 예측했고 실제로 0인 애들 14개 
# 0은 클래스라고 보면 된다. 숫자라기보다 그래서 0이 위에 있는 것
# 모델이 2라고 예측했는데 진짜는 1이었던게 하나 있고 나머지는 2에 13개가 있었다.
# 하나만 틀리고 다 맞춤 
# 이를 혼동행렬이라고 한다.

# 지도학습에서만 사용가능한거죠? => y
# y가 있을 때에만 사용 가능하다.
```

```python
# average default : binary(이진 분류)
# precision : True라고 분류한 것들 중 정답인 비율
# 0 => 모델이 14개 정답이라고 했는데 실제 정답인 애들은 = 1  14/14
# 1 => 실제 정답인 애들은 = 1  17/17=1
# 2 => 13/14  (비율로 따지기)  0.9285714285714286
# 정답 비율  1이 max  0~1인것?
precision = precision_score(y_test, predict, average=None)
for target, score in zip(iris.target_names, precision):
    print(f"{target}의 정밀도 : {score}")
```

```python
#  recall : 실제 True 중 true로 예측한 비율
recall = recall_score(y_test, predict, average=None)
for target, score in zip(iris.target_names, recall):
    print(f"{target}의 재현률 : {score}")
# 이번엔 가로로 보기 row로
# 0 => 14/14 = 1
# 1 => 17/18 = 0.9444444444444444
# 2 => 13/13 = 1
# 정밀도 어떻게 평가할건데?
```

```python
# 정밀도 precision => model 기준에서 예측한것 기준으로 진짜 맞춘거 보자(세로로 보기)
# 재현률 recall => 전체 정답 중에 정답을 맞춘 애들(가로로 보기)
```

```python
# 만일 데이터가 불균형할 때는 어떻게 평가해야 될까?
# 예를 들어 95개는 맞는데 5개의 오류가 있는 것을 95%의 정확성이라고 하고 학습을 시키면 얘도 이걸 정상이라고 판별할 것
# precision과 recall의 조화평균

# 2 * ((precision * recall) / (precision + recall))
f1 = f1_score(y_test, predict, average=None)
for target, score in zip(iris.target_names, f1):
    print(f"{target} 의 f:{score}")
```

```python
accuracy = accuracy_score(y_test, predict)
print(accuracy)
# 정확도가 뭐죠? => 전체에서 정답인 비율
# accuracy 전체에서 정확도가 맞춘 정도
```

```python
report = classification_report(y_test, predict, output_dict=True)
report
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(report)
df.rename(columns=dict(zip(["0", "1", "2"], iris.target_names)), inplace=True)
df
```

```python
# 예측 확률(클래스에 대한 출력 : 0~1 사이, 출력의 총 합은 1)
proba = model.predict_proba(x_test)
proba
# [0] = 0일 확률이 0.9 일거야 라고 하는 것
#       0일 확률 0.9(90%)  1일확률 0.01   2일확률 0.008
```

```python
proba.sum(axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(len(iris.target_names)):
    precision, recall, thresholds = precision_recall_curve(y_test, proba[:, int(i)], pos_label=i)
#     precision_recall_curve은 tpr tpr과 밀접한 관련이 있다.
    plt.xlabel("threshold")
    plt.ylabel("score")
    
    plt.plot(thresholds, precision[:len(thresholds)], "red", linestyle="--", label="precision")
    plt.plot(thresholds, recall[:len(thresholds)], "blue", label="recall")
    
    plt.legend()
    plt.title(iris.target_names[i])
    plt.show()
    
# 임계값이 올라가면 tr이 올라가고(오른쪽으로) 임계값이 내려가면 fr이 내려간다(왼쪽으로).
```

```python
# ROC curve 
# (Receiver Operating Chracteristic)

# AUC
# (Area Under the ROC Curve)  ROC Curve 아래 구역

for i in range(len(iris.target_names)):
    fprs, tprs, thresholds = roc_curve(y_test, proba[:, i], pos_label=i)
    
    plt.plot(fprs, tprs, label="roc curve")
    
    # auc가 0.5일 때는 직선(분류 못하고 있다.)
    plt.plot([0, 1], [0, 1], "--")
    
    plt.title(iris.target_names[i])
    
    plt.xlabel("tpr")
    plt.ylabel("fpr")
    plt.legend()
    plt.show()
    
# 노란색 선에 가까울수록 나빠요, 노란선 반대로 갈수록 좋아요
# 전체 면적 AUC라고 한다.
```

임계값(threshold)을 변화시키면 fpr이 변하게 된다?

임계값이 높으면 정밀도가 높아지고 fp가 낮아지므로 fpr은 0;

임계값이 낮으면 fp가 높아지고 tn은 0이므로 fpr은 1

즉 임계값을 낮추면 더 많은 항목이 양성으로 분류되므로 fpr과 tpr이 모두 증가
