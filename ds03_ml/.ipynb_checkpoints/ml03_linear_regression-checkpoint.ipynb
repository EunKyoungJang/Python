```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost(x, y, w):
    c = 0
    for i in range(len(x)):
        # 예측값 (hypothesis)
        hx= w * x[i]  #가설
        loss = (hx - y[i])**2  #오차의 제곱
        c += loss  
        
        # mse 오차의 제곱을 평균해서
        return c/len(x)
```

```python
x = [1, 2, 3]
y = [1, 2, 3]
```

```python
print(f"w:-1 / cost: {cost{x, y, -1}}")
print(f"w:0 / cost: {cost{x, y, 0}}")
print(f"w:1 / cost: {cost{x, y, 1}}")
print(f"w:2 / cost: {cost{x, y, 2}}")
print(f"w:3 / cost: {cost{x, y, 3}}")
```

```python
for i in range(-30, 50):
    w = i / 10
    c = cost(x, y, w)
    plt.plot(w, c, "ro")
    
plt.title("cost function")
plt.xlabel("weight")
plt.ylabel("cost")
plt.show()
```

```python
def gradient_descent(x, y, w):
    c = 0
    for i in range(len(x)):
        hx = w * x[i]
        loss = 2 * x[i] * (hx - y[i])
        """
        cost(w) = (w * x[i] - y[i])^2 -> 미분  # (w * x[i] - y[i])^2 
        cost(w) = w^2 * x[i]^2 - 2 * w * x[i] * y[i] + y[i]^2
        cost(w)'= 2 * w^2 * x[i] - 2 * w * y[i]
                = 2 * x[i] * (w * x[i] - y[i])
                = 2 * x[i] * (hx - y[i])
        """
        # (w * x[i] - y[i])^2  는 오차 mse
        c += loss
        
#         기울기를 다 구해서 평균을 낼게요 = 결국은 평균
        return c / len(x)
```

```python
# 기울기를 구한다는 이야기는 뭘까? 미분으로 w를 구할 때 오차가 어떻게 나오는지 보겠다.
```

```python
w = 10
old = 0
for i in range(1000):
    c = cost(x, y, w)
    grad = gradient_descent(x, y, w)
#     gradient_descent(x, y, w)에서 나오는 결과 값이 0이 되는 것을 찾는 것
    w -= 0.1 * grad
    print(f"[{i:03d}]cost:{c} old: {old} weight:{w}")
    
    #ealry stopping
    if c == old:
    # 이전 상태에서 변하지 않는다면
        break
        
    old = c
print(f"weight:{w} train: {i+1}회")

# cost => mse 
# cost가 떨어진다 => 오차가 줄어든다.
# 우리가 알고 있는 w(weight) 값이 점점 정답에 근첩한 값을 찾고 있다.
```
