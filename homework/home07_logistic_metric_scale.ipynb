```python
import pandas as pd
```

```python
df = pd.read_csv("../datasets/weight_height.csv", encoding="euc-kr")
df
```

```python
df.columns
```

```python
df = df[["학교명", "학년", "성별", "키", "몸무게"]]
df
```

```python
df["grade"] = list(map(lambda x: 0 if x[-4:] == "초등학교" 
                      else 6 if x[-3:] == "중학교"
                      else 9, df["학교명"])) + df["학년"]
df
```

```python
df.drop(["학교명", "학년"], axis='columns', inplace=True)
df
```

```python
df.columns = ["gender", "height", "weight", "grade"]
df
```

```python
# 데이터를 스케일링 후 

# 몸무게와 키와 성별을 학습시켜서
# 몸무게와 키로 성별을 예측하자.(분류)
# 남자:0, 여자:1 형태로 변환해야할껄?
# 평가지표 사용해보기!(분류할 때 평가지표들) -> confusion matrix, precision, recall, roc_curve, roc_auc_score
```

```python
import numpy as np

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, precision_score, recall_score, f1_score, accuracy_score, classification_report, precision_recall_curve, roc_curve, roc_auc_score 
```

```python
df["gender"] = df["gender"].map(lambda x: 0if x == "남" else 1)
df
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
# 1. 데이터 준비
# x = 독립변수 문제
# y = 종속변수 문제의 답
x = df[["weight", "height"]] 
y = df[["gender"]]
```

```python
x_scale = StandardScaler().fit_transform(x)
```

```python
# 2. 데이터 분할
# 테스트 세트 -> 제대로 된 모델이 만들어졌는지 확인하기 위해서
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# 3. 모델 선택
model = LogisticRegression()
```

```python
# 4. 학습
model.fit(x_train, y_train)
```

```python
#  5. 평가
model.score(x_test, y_test)
```

```python
#  6. 예측
predict = model.predict(x_test)
predict
```

```python
cm = confusion_matrix(y_test, predict)
cm
```

```python
ConfusionMatrixDisplay(cm).plot
# 0 남  1 여 였음
```

TP : True(정답) Positive(정답 예측)  : 여학생 예측 -> 여학생 맞아
    
TN : True(정답) Negative(오답 예측)  : 남학생 예측 -> 남학생 맞아
       
FP : False(오답) Positive(정답 예측) : 여학생 예측 -> 남학생이야
    
FN : False(오답) Negative(오답 예측) : 남학생 예측 -> 여학생이야



Precision = TP / (TP + FP) : True로 분류한 것 중 정답   (완정정확하게 따지기)
                             여학생이라고 예측했는데 진짜 여학생인 비율
                             실제 여학생 / 여학생으로 예측(근데 실제 남자도 포함되있음)
                             ^^ (여학생 예측 중에 진짜 여학생)
                             일단은 여학생이라고 예측한 애들 중 진짜 여학생의 비율

Recall = TP / (TP + FN) : 실제 True 중 True로 예측한 비율  (적당히 소프트하게 따지기)
                          실제 여학생 / 예측시 여학생이라고 분류해둔 전체 여학생 
                          여학생이라고 예측했는데 진짜 여학생 + 남학생이라고 예측했는데 사실은 여학생 
                          여학생이라 예측한 여학생 = 실제여학생
                          
                          

TPR : True Positive Rate : TP / (TP + FP) -> Recall
(맞았다고 생각했는데 실제로 맞았던 애들의 비율)

FPR : False Positive Rate : FP / (FP + TN) 
(맞았다고 생각했는데 사실은 틀렸던 애들의 비율)

```python
TN, FP, FN, TP = cm.ravel()
girl_precision = TP / (TP + FP)
girl_recall = TP / (TP + FN)
print(girl_precision)
print(girl_recall) 
```

```python
precision = precision_score(y_test, predict)
recall = recall_score(y_test,  predict)
print(precision)
print(recall)
```

```python
my_f1_score = 2 * ((girl_precision * girl_recall) / (girl_precision + girl_recall)) 
my_f1_score
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, proba[:, 1], pos_label=1)

plt.xlabel("threshold")
plt.ylabel("score")

plt.plot(thresholds, precision[:len(thresholds)], "red", linestyle="--", label="precision")

plt.plot(thresholds, recall[:len(thresholds)], "blue", label="recall")

plt.legend()
plt.show()

# 오른쪽으로가면(올라가면) precision(빨강)은 올라가고 recall(파랑)은 내려간다
# 이는 결국 여학생을 잘 골라낸다는 뜻
```

threshold 기준

높아지면 : precision 높아짐 (예측한 여학생에서 진짜 여학생 비율) + 정밀도가 높아짐
            -> FP 낮아짐(예측을 했는데 사실은 남학생이야)  
            -> FN 높아짐(남학생이라고 예측했는데 사실은 여학생인 비율)
            여학생을 여학생이라고 맞게 판단한게 더 높아진다. 오류비율이 떨어지는 것
            -> FPR 낮아짐 
            (여학생이라고 생각했는데 사실은 남학생이었어 라는 비율이 떨어지는 것)
            

낮아지면 : recall 높아짐 (전체 여학생에서 실제 여학생의 비율이 높아진다.)
                          예측시 여학생이라고 분류한 것에서 실제 여학생 비율
           -> FP 높아짐 -> FPR 높아짐
           
           
1) 여학생을 뽑아서 여대로 진학시킬거다. 이럴 때 threshold를 어디로 보내야할까요?
-> 오른쪽 (FP 낮아지도록 : 사실은 남학생인 비율을 줄이자)  
    진짜 여학생이 높은거


2) 여학생에게 립스틱을 지원해줄거다. threshold를 어디로 보내야할까?
-> 왼쪽 (전체 여학생들이 최대한 받으라고 : 남학생들이 받아도 괜찮은 경우)
    예측한 여학생이 높은거

진짜만 고심해서 선별해야하는지,
